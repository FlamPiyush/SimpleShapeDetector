```python
import numpy as np
import cv2
import random

vid = cv2.VideoCapture(0)
shapes = []
check_shapes = {
   "tri":0,
    "square" : 0,
    "pent" :0,
    "hex":0,
    "sept" :0,
    "oct" :0,
    "nana":0,
    "dec":0
}
no_sides = {
    "tri":11,
    "square" : 12,
    "pent" :14,
    "hex":6,
    "sept" :7,
    "oct" :8,
    "nana":9,
    "dec":10
}
name_shape = {v: k for k, v in no_sides.items()}
i = 0
start_rec = False
while(True):
      
    # Capture the video frame
    # by frame
    ret, img = vid.read()
    var = cv2.waitKey(1)
    if  var & 0xFF == ord('r'):##start recording
        start_rec = True
        imgGrey = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    #     imgGrey = cv2.blur(imgGrey,(3,3))
    #     _, thrash = cv2.threshold(imgGrey, 175, 255, cv2.THRESH_BINARY)
        thrash = cv2.Canny(imgGrey,100,200)
    #     kernel = np.ones((5,5), np.uint8)
    #     thrash = cv2.dilate(thrash,kernel,iterations=1)
    #     thrash = cv2.erode(thrash,kernel,iterations=1)
        contours, _ = cv2.findContours(thrash, cv2.RETR_TREE, cv2.CHAIN_APPROX_NONE)
        closed_contours = []
        for i in contours:
            if cv2.contourArea(i) > cv2.arcLength(i, True):
                closed_contours.append(i)

        cv2.imshow("img", thrash)
        for contour in closed_contours:
            i = random.randrange(0, 255, 20)
            j = random.randrange(0, 255, 20)
            k = random.randrange(0, 255, 20)
            approx = cv2.approxPolyDP(contour, 0.01* cv2.arcLength(contour, True), True)
            cv2.drawContours(img, [contour], 0, (i, j, k), 2)
            x = approx.ravel()[0]
            y = approx.ravel()[1] - 5
            try:
                if name_shape[len(approx)] not in shapes:
                    shapes.append(name_shape[len(approx)])
            except:
                print(None)
#             elif len(approx) == 14:
#                 cv2.putText(img, "Pentagon", (x, y), cv2.FONT_HERSHEY_COMPLEX, 0.5, (255, 0, 0))
#                 if "pentagon" not in shapes:
#                     shapes.append("Pentagon")
#             elif len(approx) == 6:
#                 cv2.putText(img, "Hexagon", (x, y), cv2.FONT_HERSHEY_COMPLEX, 0.5, (255, 0, 0))
#                 if "Hexagon" not in shapes:
#                     shapes.append("Hexagon")
#             elif len(approx) == 7:
#                 cv2.putText(img, "Heptagon", (x, y), cv2.FONT_HERSHEY_COMPLEX, 0.5, (255, 0, 0))
#                 if "Heptagon" not in shapes:
#                     shapes.append("Heptagon")
#             elif len(approx) == 8:
#                 cv2.putText(img, "Octagon", (x, y), cv2.FONT_HERSHEY_COMPLEX, 0.5, (255, 0, 0))
#                 if "Octagon" not in shapes:
#                     shapes.append("Octagon")
#             elif len(approx) == 11:
#                 cv2.putText(img, "11gon", (x, y), cv2.FONT_HERSHEY_COMPLEX, 0.5, (255, 0, 0))
#                 if "11gon" not in shapes:
#                     shapes.append("11gon")
        print("shapes::::", shapes)
        for i in shapes:
            check_shapes[i] +=1
        shapes = []
       
        
    

    cv2.imshow("shapes", img)
    if var & 0xFF == ord('q'):
        break
    
    
# After the loop release the cap object
vid.release()
cv2.destroyAllWindows()
```

```python
codeConversion = {
    "0":"tri",
    "1":"square",
    "2":"pent",
    "3":"hex",
    "4":"sept",
    "5":"oct",
    "6":"nana",
    "7":"dec"
}
codeConversion_inv = {v: k for k, v in codeConversion.items()}
```

```python
threshold = check_shapes[max(check_shapes, key=check_shapes.get)]/3
```

```python
answer = "00000000"
```

```python
for key in check_shapes.keys():
    if(check_shapes[key]>threshold):
        answer = answer[0:int(codeConversion_inv[key])] + '1' + answer[int(codeConversion_inv[key])+1: ]
```
